```python
# Data Cleaning
# Check for missing values
print(user_details.isnull().sum())
print(cooking_sessions.isnull().sum())
print(order_details.isnull().sum())
```

So in the order_details table we have 2 null rows in the rating column. Let's take a look at those rows.

```python
null_rows = order_details[order_details['Rating'].isna()]

print(null_rows)
```

```python
# Merge datasets
merged_data = order_details.merge(cooking_sessions, on=["User ID", "Session ID"], how="left")
merged_data = merged_data.merge(user_details, on="User ID", how="right")

merged_data
```

It seems that we don't have much information about the user id 9 and 10 even they have placed some orders after there registration date.

```python
# 1. Popular dishes
popular_dishes = merged_data.groupby("Dish Name_x")["Order ID"].count().sort_values(ascending=False)
print("Popular Dishes:\n", popular_dishes.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Visualization
plt.figure(figsize=(10, 6))
# Popular Dishes Bar Plot
sns.barplot(x=popular_dishes.head(10).index, y=popular_dishes.head(10).values)
plt.title("Popular Dishes")
plt.xlabel("Dish Name")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
plt.show()
```
